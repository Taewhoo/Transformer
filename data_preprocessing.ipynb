```python
from datasets import load_dataset
dataset = load_dataset("wmt14",'de-en')
```

```python
dataset
```

```python
train = dataset['train']
val = dataset['validation']
test = dataset['test']
```

```python
def create_paired_list(dataset):
    paired_list = []
    pairs = dataset['translation']
    for pair in pairs:
        paired_list.append([pair['en'], pair['de']])
    
    return paired_list
```

```python
train_pairs = create_paired_list(train)
val_pairs = create_paired_list(val)
test_pairs = create_paired_list(test)
```

```python
len(train_pairs)
```

```python
import pandas as pd

train_df = pd.DataFrame(train_pairs[:250000], columns=['src', 'tar'])
train_df
```

```python
train_df.to_csv("train_df_250k.tsv", sep='\t')
```

```python
from IPython.display import FileLink
FileLink("train_df_250k.tsv")
```

```python
import random
random.seed(42)

src_list = train_df['src'].tolist()
tar_list = train_df['tar'].tolist()
total_list = src_list + tar_list

random.shuffle(total_list)
print(total_list[:5])

with open('train_shuffled.txt', 'w') as wf:
    wf.write('\n'.join(total_list))            
```
